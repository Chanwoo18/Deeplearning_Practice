```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from matplotlib import pyplot
from sklearn.preprocessing import StandardScaler


df = pd.read_csv('./01.data/01.student/HT_Student_mat_Train_20230511.csv')
df
```

```python
# 세부 정보를 X로 지정합니다.
X = df.iloc[:,0:32]
# 당뇨병 여부를 Y로 지정합니다.
y = df.iloc[:,32]
```

```python
#범주형 데이터 원핫 인코딩
X = pd.get_dummies(X, columns=['school', 'sex', 'address', 'famsize', 'Pstatus',
                                 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian',
                                 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
                                 'higher', 'internet', 'romantic'])
X
```

```python
# 실제 예측할 때 원핫인코딩 된 컬럼의 순서를 저장
feature_names = X.columns.tolist()
```

```python
import pandas as pd

# 학습셋과 테스트셋을 75:25의 비율로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)


### 2. 랜덤 포레스트

# 랜덤 포레스트 분류기를 불러 옵니다.

from sklearn.ensemble import RandomForestClassifier


# 학습 환경을 설정합니다.
classifier= RandomForestClassifier(n_estimators = 50)
classifier.fit(X_train, y_train)

# 테스트셋에 적용합니다.
y_pred = classifier.predict(X_test)

# 계층별 교차 검증 환경을 설정합니다.
skf=StratifiedKFold(n_splits=10, shuffle=True)

# 교차 검증을 통해 정확도를 계산합니다.
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = skf)

# 정확도와 표준편차를 출력합니다.
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))


#MAE 값을 출력합니다.
from sklearn.metrics import mean_absolute_error

print(f'MAE의 값 : {mean_absolute_error(y_test, y_pred)}')
```
