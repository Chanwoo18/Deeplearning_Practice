```python
import pandas as pd

#자세에 대한 데이터를 불러옵니다.
df = pd.read_csv('./dataaa/02.HAR/HT_HAR_112_Train_20230511.csv')
```

```python
df
```

```python
# 정답 레이블에 대한 범주를 확인합니다.
df['Class'].unique()
```

```python
X = df.iloc[:,0:112]
y = df.iloc[:,112]

print(X[0:5])
print(y[0:5])
```

```python
# 정답 레이블에 대해 원핫 인코딩
y = pd.get_dummies(y)
print(y[0:10])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 모델 설정
model = Sequential()
model.add(Dense(130,  input_dim=112, activation='relu'))
model.add(Dense(65,  activation='relu'))
model.add(Dense(32,  activation='relu'))
model.add(Dense(16,  activation='relu'))
model.add(Dense(6, activation='softmax'))
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
history=model.fit(X, y, epochs=50, batch_size=12)
```

```python
# 모델을 저장 합니다.
model.save('./test')
```

```python
from tensorflow import keras

# 모델을 불러옵니다.
model = keras.models.load_model('./test')
```
